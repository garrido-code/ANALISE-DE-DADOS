# while

faça um programa que escreva numero pares até 50

```python
#codico ex 4

numero = 1
while numero <= 50 :
    if numero % 2 == 0:
        print(numero)
    numero += 1 
```

Faça um programa que conte de 10 até 0 (de trás para frente)
